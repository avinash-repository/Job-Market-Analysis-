```python
import pandas as pd 
import numpy as np




df = pd.read_csv(r'E:\Data Analyst\Project\Scraped Dataset\Job Analysis\expanded_job_market.csv')
pd.set_option('display.max_rows' , None)
df['Job Type'] = np.where(
    df['Job Title'].str.contains('virtual|remote|work from home', case=False, na=False) |
    df['Location'].str.contains('virtual|remote|work from home', case=False, na=False),
    'Remote',
    'On-site'
)


df = df.drop_duplicates()
df = df.dropna()

df['Location'] = df['Location'].str.replace(r'\b(Calcutta|Howrah)\b' , 'Kolkata' , regex=True)

df['Job Title'] = df['Job Title'] \
    .str.replace(r'\b\d{1,2}(am|pm)\s*-\s*\d{1,2}(am|pm)\b', '', regex=True) \
    .str.replace(r'\([^)]*\)', '', regex=True) \
    .str.replace(r'\bany time\b', '', regex=True, case=False) \
    .str.strip()

# Remove all special characters from both ends and capitalize
df['Job Title'] = df['Job Title'] \
    .str.replace(r'^[^\w]+|[^\w]+$', '', regex=True) \
    .str.title()  
df['Job Title'] = df['Job Title'].str.replace(r'\$\d+(\.\d+)?\s*-\s*\$\d+(\.\d+)?', '', regex=True).str.strip()


df['Industry'] = df['Industry'].str.replace(r'\([^)]*\)' , '' , regex=True)

df['Salary'] = df['Salary'] \
      .str.replace(r'\([^)]*\)', '', regex=True)

    
#Function to remove Country name from job title 
def replace_country_name(name):
    if name['Location'] in name['Job Title']:
        return name['Job Title'].replace(name['Location'] , '')
    return name['Job Title']   

df['Job Title']= df.apply(replace_country_name , axis=1)
```

```python
import re


conversion_rates = {
    'INR': 0.012,  
    'IDR': 0.000065, #indoneasia
    'CAD': 0.75,    #canadian dollar
    'AUD': 0.65,    #australian dollar 
    'GBP': 1.25,  #pound 
    'EUR': 1.1, #euro
    'T': 1000,  #thousands
    'L': 100000, #lakhs
}



def convert_currency(amount, unit):
    if unit in conversion_rates:
        return round(amount * conversion_rates[unit], 2)
    return amount  


def process_salary(salary_val):
    currency_map = {'₹': 'INR', '$': 'USD', 'A$': 'AUD', '€': 'EUR', 'C$': 'CAD', '£': 'GBP'}

    if 'T' in salary_val or 'L' in salary_val:
        match_result = re.search(r'(\d+)([TL])', salary_val)
        if match_result:
            amount = int(match_result.group(1))
            unit = match_result.group(2)
            final_amount = convert_currency(amount, unit)
            return f"${round(final_amount / 2080, 2)} per hour"  

    match = re.search(r'([₹$A€C£])?(\d+(?:[.,]\d+)?)', salary_val)
    if match:
        currency_symbol = match.group(1) if match.group(1) else '$'
        amount = float(match.group(2).replace(',', ''))  

        currency = currency_map.get(currency_symbol, 'USD') # Default to USD if no match
        final_amount = convert_currency(amount, currency)

        if final_amount > 800:
            hourly_rate = final_amount / 2080
            return f"${round(hourly_rate, 2)} per hour"
        
        if 'hour' in salary_val.lower():
            return f"${round(final_amount, 2)} per hour"
        elif 'month' in salary_val.lower():
            # Directly divide by 160 to get hourly rate from monthly
            return f"${round(final_amount / 160, 2)} per hour"
        else:
            return f"${round(final_amount, 2)} per hour"
    
    return "Unable to process"


# Apply function to Salary column in DataFrame
df['Salary'] = df['Salary'].apply(process_salary)
df['Salary'] = df['Salary'].str.replace('per hour', '')
df.head(100)
```
